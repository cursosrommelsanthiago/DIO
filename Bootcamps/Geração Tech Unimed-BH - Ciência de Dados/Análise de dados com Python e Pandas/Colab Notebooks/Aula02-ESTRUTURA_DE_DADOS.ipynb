# ESTRUTURA DE DADOS COM ***PYTHON***

Criando lista

```python
animais = [1,2,3]
print(animais)
```

Inserindo items na lista

```python
animais = ["cachorro", "gato", 12345, 6.5]
print(animais)
```

Imprimindo itens de acordo com sua posição

```python
print(animais[0])
print(animais[3])
```

### Manipulando a lista com os metodos de adicionar, substituir, remover e buscar

```python
animais[0] = "papagaio"
print(animais)
```

```python
animais.remove("gato")
print(animais)
```

```python
print(len(animais))
```

```python
print("gato" in animais)
```

```python
lista = [500, 30, 300, 80, 10]
```

```python
print(max(lista))
```

```python
print(min(lista))
```

```python
animais.append(["leão", "Cachorro"])
print(animais)
```

```python
animais.extend(["cobra", 6])
print(animais)
```

```python
print(animais[3].count("leão"))
```

```python
lista.sort()
print(animais)
```
